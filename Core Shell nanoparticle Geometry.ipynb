```python
%matplotlib notebook
import numpy as np
pikseler = 5001 # Bruk oddetall
width = 100 # Varier dette tallet for bredden til shellet

centrex = centrey = (pikseler - 1) // 2 # midten av en pikseler * pikseler matrise, må være heltall
r1 = int(0.9 * pikseler/2) # Bruk 90% av bildet til partikkelen
r2 = r1 - width


y,x = np.ogrid[-centrex:pikseler-centrex, -centrey:pikseler-centrey] # Denne lager en matrise (array)
mask1 = x*x + y*y <= r1*r1

array = np.zeros((pikseler, pikseler))
array[mask1] = 1

mask2 = x*x + y*y <= r2*r2
array[mask2] = 0
```

```python
# Importer Matplotlib for å plotte
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.imshow(array)
```

```python
plt.figure()
line = array.sum(1) # Summer bildet langs med y-aksen på bildet
plt.plot(line)
```

```python
from scipy.ndimage import gaussian_filter
```

```python
filtered_line = gaussian_filter(line, sigma=100) # varier sigma for å "blurre" bildet, som er naturlig når partikkelen beveger seg.
```

```python
plt.figure()
plt.plot(filtered_line)
```
